**Methods and Results:**



The susceptible population's rate of change (dS/dt) is calculated as follows: dS/dt = -b* S * I.
Here, denotes the typical number of contacts between a susceptible person and an infected person per unit of time. It also considers the likelihood of disease transmission during each contact. The rate at which susceptible people get infected as a result of contact with infected people is denoted by the expression b* S * I.

The infected population's rate of change (dI/dt) is calculated as follows: dI/dt = b* S * I - k*S* I

The first term, b* S * I, denotes the rate of infection among those who are susceptible (as previously mentioned). The pace at which sick people recover or are expelled from the infected compartment is represented by the second term, k*S* I.  is the inverse of the average and shows the recovery rate.8The percentage of people who leave the infected compartment as a result of recovery or other circumstances is indicated by the symbol *I.

The recovered population's rate of change (dR/dt) is given by dR/dt =k*S * I.

Recovery from the infection is the sole means to access the recovered compartment. As a result, the population's rate of change is equal to the rate of individual recovery, which is k*S * I.

We optimized the SIR model using Euler’s method in order to extrapolate on the data we obtained from research. Euler’s method is a numerical method for approximating the solution of ordinary differential equations (ODEs). It entails breaking up the interval of interest into manageable stages and estimating the solution's estimated value based on the slope of the ODE at each step.
